Predict the price of a used car depending on its specification

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#loading the raw data
raw_data = pd.read_csv('1.04.+Real-life+example.csv')
raw_data.head()
```

Data Cleaning

```python
#Exploring descriptive statitics
raw_data.describe(include = 'all')
```

```python
#Dropping the Model column
data = raw_data.drop(['Model'],axis=1)
data
```

```python
#Dealing with missing values
data.isnull().sum()
```

```python
#Deleting missing values since they are less than 5% of the orignal value
data_no_mv = data.dropna(axis=0)
```

```python
data_no_mv.describe(include='all')
```

```python
#Exploring the probability distribution function. Optimally, this should be a normal distribution
sns.displot(data_no_mv['Price'], kde=True),sns.displot(data_no_mv['Mileage'], kde=True)
sns.displot(data_no_mv['EngineV'], kde=True),sns.displot(data_no_mv['Year'], kde=True)
```

```python
#Outliers are observations that lie at abnormal distances from other observations in the data, they affect the regression
#The best way to deal with outliers is remove the top 1% of observations using the quantile method
q = data_no_mv['Price'].quantile(0.99)
data_1 = data_no_mv[data_no_mv['Price']<q]
data_1.describe(include='all')
```

```python
sns.displot(data_1['Price'], kde=True)
```

```python
q = data_1['Mileage'].quantile(0.99)
data_2 = data_1[data_1['Mileage']<q]
data_2.describe(include='all')
```

```python
sns.displot(data_2['Mileage'], kde=True)
```

```python
data_3 = data_2[data_2['EngineV']<6.5]
data_3.describe(include='all')
```

```python
sns.displot(data_3['EngineV'], kde=True)
```

```python
q = data_3['Year'].quantile(0.01)
data_4 = data_3[data_3['Year']>q]
data_4.describe(include='all')
```

```python
sns.displot(data_4['Year'], kde=True)
```

```python
data_cleaned = data_4.reset_index(drop=True)
```

```python
data_cleaned.describe(include='all')
```

Checking OLS Assumptions

```python
#Checking for Linearity using Scatterplot
f, (ax1, ax2, ax3) = plt.subplots(1,3, sharey=True, figsize=(15,3))

ax1.scatter(data_cleaned['Year'], data_cleaned['Price'])
ax1.set_title('Price and Year')

ax2.scatter(data_cleaned['EngineV'], data_cleaned['Price'])
ax2.set_title('Price and EngineV')

ax3.scatter(data_cleaned['Mileage'], data_cleaned['Price'])
ax3.set_title('Price and Mileage')

plt.show()
```

```python
#The above scatterplots did not yeild any linear pattern, transformation of the variables should be done before regression
#When dealing with such exponential scatterplots, the log transformation is preferred
log_price = np.log(data_cleaned['Price'])
data_cleaned['log_price'] = log_price
data_cleaned
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1,3, sharey=True, figsize=(15,3))

ax1.scatter(data_cleaned['Year'], data_cleaned['log_price'])
ax1.set_title('log_price and Year')

ax2.scatter(data_cleaned['EngineV'], data_cleaned['log_price'])
ax2.set_title('log_price and EngineV')

ax3.scatter(data_cleaned['Mileage'], data_cleaned['log_price'])
ax3.set_title('log_price and Mileage')

plt.show()
```

```python
#From the above scatterplots and following log transformation of the price (dependent variable),
#linear scatterplots have been generated
#Therefore, the original price data needs to be dropped from the dataframe
data_cleaned = data_cleaned.drop(['Price'], axis=1)
```

```python
#Multicolinearity using VIF (variance inflation factor)
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage','EngineV','Year']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor (variables.values, i) for i in range (variables.shape[1])]
vif['features'] = variables.columns
```

```python
#When       VIF = 1 = No multicolinearity
#           1 < VIF < 5 = Perfectly okay
#           6 < VIF < 10 = Unacceptable
#That is Year is too correlated with the other variables, therefore is should only be removed 
vif
```

```python
data_no_multico = data_cleaned.drop(['Year'], axis=1)
```

```python
data_no_multico
```

Create Dummy Variables

```python
#Create dummy variables for categorical data
data_with_dummies = pd.get_dummies(data_no_multico, drop_first=True)
```

```python
data_with_dummies.head()
```

```python
#Reorder the columns so that the dependent variable (log_price) is the first in the dataframe 
data_with_dummies.columns.values
```

```python
cols = [ 'log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed
```

Linear Regression Model

```python
#Declare the inputs and targets
targets  = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'], axis = 1)
```

```python
#Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
#Standardise the scaled data
inputs_scaled = scaler.transform(inputs)
```

```python
#Train Test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

```python
#Create the regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
#To check the final result, the predicted values is plotted against the observed values (Targets)
y_hat = reg.predict(x_train)
```

```python
#Create a scatterplot of y_train (Target) and y_hat (Predictions)
#If a 45 degree line is drawn, which shows the best possible match btw the targets and predictions,
#the close the scatterplot is to this line the better the model
#Our result is not perfect, but its not random as the points are situated around the 45 degree line
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#Checking for the distribution of the residuals using the residual plot
sns.displot(y_train - y_hat, kde=True)
plt.title('Residuals PDF', size=18)
```

```python
#Calculating the R-squared: How much of the variability of the data the model is explaining 
reg.score(x_train,y_train)
```

```python
#Calculating the intercept
reg.intercept_
```

```python
#Calculating the coefficient
reg.coef_
```

```python
#Summary table
#A positive weight shows that as a feature increases in value the log_price and 'price' increase, respectively
#A negative weight shows that as a feature increases in value the log_price and 'price' decreases, respectively
#For the dummy variable, a negative value imply that the dummy is cheaper than the benchmark(Audi) and vice versa
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

Testing

```python
#Finding the predictions
y_hat_test = reg.predict(x_test)
```

```python
#From the plot, for higer prices, there is a higher concentration of values around the 45 degree line, implying that the 
#model is good at predicting higher prices, however, for lower prices, the points are more scattered because the prices
#are not being gotten right.
#From the graph, the more saturated the color, the more the concentration of points, and vice versa. Also most of the 
#points are close to the 45 degree line
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Targets (y_train)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#To manually explore what the algorithm came up with to determine its effectiveness
df_performance = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_performance.head()
```

```python
#Putting the prices before the predictions 
y_test = y_test.reset_index(drop=True)
df_performance['Target'] = np.exp(y_test)
df_performance.head()
```

```python
#Comparing the targets and predictions using the residual or difference
df_performance['Residual'] = df_performance['Target'] - df_performance['Prediction']
df_performance['%Difference'] = np.absolute(df_performance['Residual'] / df_performance['Target'] * 100)
df_performance.head()
```

```python
pd.options.display.max_rows = 999
df_performance.sort_values(by=['%Difference']).round(2)
```

Revising this dataFrame manually we can see for which types of observations we've got good predictions,
and for which we are very far off. Going to the bottom of the dataFrame, we see that there are very few predictions 
dramatically far off from the observed values.

If you look closely at the observed column, you notice that the observed prices are extremely low, $3000,$4000 $2600.
As we know, our model is using the mileage, engine volume, brand, registration and body type to predict the price of a used 
car. On average, it is pretty decent at predicting the price.

For these last samples though it isn't. An important note is that all residuals for these outliers are negative. Therefore, 
their predictions are higher than the targets. An explanation may be that we are missing an important factor which drives 
the price of a used car lower. It may be the model of the car which we removed at the beginning of the analysis but it may 
also be that the car was damaged in some way, a piece of information we did not initially have.

There are different ways in which we can make our model better: Using a different set of variables, removing a bigger part 
of the outlier observations or using different kinds of transformations.
